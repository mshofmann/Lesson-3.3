# Box Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats


loansData = pd.read_csv('https://spark-public.s3.amazonaws.com/dataanalysis/loansData.csv')
loansData.dropna(inplace=True)
loansData.boxplot(column='Amount.Funded.By.Investors')
plt.show()
loansData.boxplot(column='Amount.Requested')
```

While both Amount Funded and Amount Requested have a similar median, the third quartile of Amount Funded is slightly lower than that of Amount Requested.

```python
FundedRange = max(loansData['Amount.Funded.By.Investors'])-min(loansData['Amount.Funded.By.Investors'])
FundedRange
```

```python
RequestedRange = max(loansData['Amount.Requested'])-min(loansData['Amount.Requested'])
RequestedRange
```

Conversely, the range of Amount Funded is larger than Amount Requested by 1000.

```python
loansData.hist(column='Amount.Funded.By.Investors')
plt.axis([0, 35000, 0, 600])
plt.show()
loansData.hist(column='Amount.Requested')
plt.axis([0, 35000, 0, 600])
```

The distribution of amounts funded and requested follow a similar curve.

```python
plt.figure()
graph = stats.probplot(loansData['Amount.Funded.By.Investors'], dist='norm', plot=plt)
plt.title('Amount Funded by Investors')
plt.show()
plt.figure()
graph2 = stats.probplot(loansData['Amount.Requested'], dist='norm', plot=plt)
plt.title('Amount Requested')
```
